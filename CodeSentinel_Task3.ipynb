## 1. Import Libraries & Load Dataset
We will load the Netflix dataset (or any dataset of choice) using pandas.


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset (replace with your file path)
df = pd.read_csv("netflix_data.csv")

# Preview dataset
df.head()
```

## 2. Correlation Matrix & Heatmap
We’ll compute correlations for numeric columns and visualize them with a heatmap.


```python
# Compute correlation (only numeric columns)
corr = df.corr(numeric_only=True)

# Plot heatmap
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix Heatmap")
plt.show()
```

## 3. Top 5 Most Frequent Values
Let’s check the most common values for selected columns (e.g., `director`, `country`, `rating`).


```python
# Show available columns
print("Available columns:\n", df.columns.tolist(), "\n")

# Safely compute top 5 most frequent values
for col in ['director', 'country', 'rating']:  # adjust list based on your dataset
    if col in df.columns:
        print(f"Top 5 {col.title()}:\n", df[col].value_counts().head(5), "\n")
    else:
        print(f"Column '{col}' not found in dataset.\n")
```

## 4. Distribution Plots
We’ll plot histograms and boxplots to understand distributions of numeric features (e.g., `release_year`).


```python
# Histogram for release year
plt.figure(figsize=(8,4))
sns.histplot(df['release_year'], bins=20, kde=True)
plt.title("Distribution of Release Year")
plt.show()

# Boxplot for release year
plt.figure(figsize=(8,4))
sns.boxplot(x=df['release_year'])
plt.title("Boxplot of Release Year")
plt.show()
```
